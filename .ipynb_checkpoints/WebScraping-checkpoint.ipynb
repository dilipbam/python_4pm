```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
soup = BeautifulSoup(soup.text, "lxml")
soup.text
```

```python
table_code = soup.table
table_code
```

```python
table_code.find_all('tr')
```

```python
final_data = []
tr_tags = table_code.find_all('tr')
for i in tr_tags:
    x = i.text.split('\n')
    if x[1] != "":
        final_data.append(x[1:-1])
# final_data
# tr_tags[0].text
```

```python
tr_tags = table_code.find_all('tr')
# for i in tr_tags:
#     print(i.text)
tr_tags[0].text.split('\n')
```

```python
# saving scrapped data into csv file
import csv
file = open('covid_Data.csv','w')
x = csv.writer(file)
x.writerows(final_data)
```

```python
import pandas as pd
df = pd.read_csv('covid_Data.csv', encoding='latin1')
df
```

```python
df.info()
```

```python
# removing comma from the totalcases column and save into list
Country = list(df['Country,Other'][0:5])
TotalCases=list(map(lambda x:int(x.replace(',','')),list(df['TotalCases'][0:5])))
TotalRecovered=list(map(lambda x:int(x.replace(',','')),list(df['TotalRecovered'][0:5])))
```

```python
import plotly.graph_objects as go

fig = go.Figure([go.Bar(x=Country, y=TotalCases)])
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[
    go.Bar(name='TotalCases', x=Country, y=TotalCases),
    go.Bar(name='TotalRecovered', x=Country, y=TotalRecovered)
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```
